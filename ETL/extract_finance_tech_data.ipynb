```python
#import dependencies
# %pip install yfinance
# %pip install pandas_datareader
# %pip install pandas
%pip install yfinance --upgrade --no-cache-dir 
import pandas as pd
import yfinance as yf
from pandas_datareader import data as wb
```

```python
#create dataframe to hold the information of Tickers
info_df = pd.DataFrame({"Ticker" : ["WD", "PFSI", "LDI", "GHI", "COOP", "AFL", "HIG", "PRU", "ALL", "PGR", "MS", "BLK", "GS", "TROW", "BEN", "V", "MA", "AXP", "DFS", "COF",
                                    "QCOM", "TXN", "ADI", "MU", "ARM", "CRM", "SAP", "INTU", "NOW", "ZI", "AAPL", "MSFT", "SONY", "LPL", "SONO", "ACN", "IBM", "FI", "INFY", "CTSH",
                                    "ISRG","BDX","ALC","WST","RMD","UNH","ELV", "CVS","CI","HUM","NVO","VRTX","REGN","AMGN","MRNA","PFE","JNJ","LLY","AZN", "MRK",
                                    "CPRT", "KMX", "PAG", "LAD", "AN", "MBLY", "APTV", "MGA", "LKQ", "ALV", "TSLA", "TM", "HMC", "GM", "F", "NVDA", "TSM", "AVGO", "AMD", "INTC"],
                        "Industry" : "",
                        "Sector" : ""})

info_df["Industry"].iloc[0:5] ="Mortgage"
info_df["Industry"].iloc[5:10] ="Insurance"
info_df["Industry"].iloc[10:15] ="Asset Management"
info_df["Industry"].iloc[15:20] ="Credit Services"

info_df["Industry"].iloc[20:25] ="Semiconductors"
info_df["Industry"].iloc[25:30] ="Software"
info_df["Industry"].iloc[30:35] ="Consumer"
info_df["Industry"].iloc[35:40] ="Infotech"

info_df["Industry"].iloc[40:45] = "Healthcare plans"
info_df["Industry"].iloc[45:50] = "Healthcare plans"
info_df["Industry"].iloc[50:55] = "Biotechnology"
info_df["Industry"].iloc[55:60] = "Drug Manufacturers"

info_df["Industry"].iloc[60:65] = "Auto & Truck Dealerships"
info_df["Industry"].iloc[65:70] = "Auto Parts"
info_df["Industry"].iloc[70:75] = "Auto Manufacturers"
info_df["Industry"].iloc[75:80] = "Semiconductors"

info_df["Sector"].iloc[0:20] = "Finance"
info_df["Sector"].iloc[20:40] = "Technology"
info_df["Sector"].iloc[40:60] = "Healthcare"
info_df["Sector"].iloc[60:80] = "Automobile"

#Display dataframe
info_df
```

```python
# List to temporarily hold data
ls = []

# DataFrame to hold all retrieved data
final_df = pd.DataFrame()

# Loop through each ticker in info_df
for sym in info_df["Ticker"]:
    # Create a Ticker object
    ticker = yf.Ticker(sym)

    # Fetch historical market data for the past 5 years
    stock_data = ticker.history(period="5y")  

    # Add a Ticker column for reference
    stock_data["Ticker"] = sym

    # Append data to list
    ls.append(stock_data)

# Concatenate all data into final_df
final_df = pd.concat(ls, ignore_index=False)

# Show the final dataframe
print(final_df)
```

```python
#merging the dataframes to show sector and industry
finance_tech_df = pd.merge(final_df, info_df, on = "Ticker", how = "left")
finance_tech_df["Date"] = final_df.index
finance_tech_df = finance_tech_df[["Date", "Open", "High", "Low", "Close", "Volume", "Ticker", "Industry", "Sector"]]

#show the result
finance_tech_df
```

```python
final_df = final_df.reset_index()
final_df["Date"] = pd.to_datetime(final_df["Date"]).dt.date
```

```python
final_df = final_df.drop(columns=["Dividends", "Stock Splits"])
final_df
```

```python
final_df.to_csv("Datasets/CompleteData.csv", index = False)
```
